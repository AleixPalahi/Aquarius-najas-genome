# Isotools data prep / transcriptome reconstruction

```python
# preperation: import the libraries
from isotools import Transcriptome
from isotools import __version__ as isotools_version
import pandas as pd
import matplotlib.pyplot as plt
import logging
#Set up logging
logging.basicConfig(format='%(levelname)s:%(message)s', level=logging.INFO)
logger=logging.getLogger('isotools')
logger.info(f'This is isotools version {isotools_version}')

path='/home/ingomue/1.Work/1.Projects/2025_Aquarius/15.isotools_final'
```

```python
annotation=f'/home/ingomue/1.Work/1.Projects/2025_Aquarius/05.isotools/00.input/01.ref/A_najas_filt_final_sorted.gff3.gz'
# create the IsoTools transcriptome object from the reference annotation
isoseq=Transcriptome.from_reference(annotation)
```

```python
import os

samples = ['I1', 'I2', 'I3', 'I4', 'I5F', 'I5M', 'Fad', 'Mad']
# Dictionary to store all renaming maps
all_id_maps = {}

# Loop through each sample
for sample in samples:
    print(f"\n{'='*60}")
    print(f"Adding sample: {sample}")
    print(f"{'='*60}")
    
    # Define file paths for this sample
    coverage_file = f'{path}/00.input/{sample}_abundance.csv'
    gtf_file = f'{path}/00.input/{sample}_custom_final.filtered.gtf'
    
    if not os.path.exists(coverage_file):
        print(f"  WARNING: Coverage file {coverage_file} not found, skipping")
        continue
    if not os.path.exists(gtf_file):
        print(f"  WARNING: GTF file {gtf_file} not found, skipping")
        continue
    
    # Define sample_cov_cols for this sample
    # Maps isotools sample name → column name in CSV
    sample_cov_cols = {sample: f'{sample}_coverage'}
    
    try:
        # Add this sample to the isoseq object
        id_map = isoseq.add_sample_from_csv(
            coverage_csv_file=coverage_file,
            transcripts_file=gtf_file,
            transcript_id_col='transcript_id',
            sample_cov_cols=sample_cov_cols,
            reconstruct_genes=False,
            infer_genes=True,
            sep='\t'
        )
        
        # Store the renaming map
        all_id_maps[sample] = id_map
        print(f"  Successfully added {sample}")
        print(f"  Genes renamed: {len(id_map) if id_map else 0}")
        
    except Exception as e:
        print(f"  ERROR adding {sample}: {e}")
        continue

# After adding all samples, verify
print(f"\n{'='*60}")
print("SAMPLES IN ISOSEQ OBJECT:")
print(f"{'='*60}")
print(f"Samples: {isoseq.samples}")
print(f"Number of samples: {len(isoseq.samples)}")

# the sample table of the transcriptome object contains the number of imported reads
isoseq.sample_table
```

```python
# compute qc metrics
genome_fn=f'/home/ingomue/1.Work/1.Projects/2025_Aquarius/05.isotools/00.input/01.ref/Anajas_hardmasked_genome.fa'
isoseq.add_qc_metrics(genome_fn)
isoseq.make_index()
```

```python
# export the transcriptome object for later use.
isoseq.save(f'{path}/01.output/A_najas_isotools_filtered.pkl')
# to load the data in the next session, use
# isoseq=Transcriptome.load('A_najas_isotools_substantial_isotool.pkl')
```
